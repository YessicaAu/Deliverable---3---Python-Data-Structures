```python
#To import the libaries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#To load in iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
#To drop species column
iris = iris.drop(labels = 'species', axis = 1)
iris.head()
```

```python
#To create new column named "sepal_sum" with sum of sepal_length and sepal_width columns
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
#To create new column named "petal_sum" with sum of petal_length and petal_width columns
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
#To display the mean, std, min, and max of the values of the iris dataframe in a new dataframe named "iris_summary"
summary_dict = [{'Mean': np.mean(iris['sepal_length']),'std': np.std(iris['sepal_length']), 'Min': np.min(iris['sepal_length']), 'Max': np.max(iris['sepal_length'])},
        {'Mean': np.mean(iris['sepal_width']), 'std': np.std(iris['sepal_width']), 'Min': np.min(iris['sepal_width']), 'Max': np.max(iris['sepal_width'])},
        {'Mean': np.mean(iris['sepal_sum']), 'std': np.std(iris['sepal_sum']), 'Min': np.min(iris['sepal_sum']), 'Max': np.max(iris['sepal_sum'])},
        {'Mean': np.mean(iris['petal_length']), 'std': np.std(iris['petal_length']), 'Min': np.min(iris['petal_length']), 'Max': np.max(iris['petal_length'])},
        {'Mean': np.mean(iris['petal_width']), 'std': np.std(iris['petal_width']), 'Min': np.min(iris['petal_width']), 'Max': np.max(iris['petal_width'])},
        {'Mean': np.mean(iris['petal_sum']), 'std': np.std(iris['petal_sum']), 'Min': np.min(iris['petal_sum']), 'Max': np.max(iris['petal_sum'])}]

iris_summary = pd.DataFrame(summary_dict, index=['sepal_length', 'sepal_width', 'sepal_sum', 'petal_length',	'petal_width',	'petal_sum'])


iris_summary
```
